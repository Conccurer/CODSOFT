# Credit Card Fraud Detection Model:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
fraud_count = data['Class'].value_counts()
fraud_rate = 100*fraud_count/data.shape[0]
fraud_data = pd.concat([fraud_count, fraud_rate], axis=1).reset_index()
fraud_data.columns = ['Class','Count','Percentage']
fraud_data
```

```python
data_fraud = data[data['Class']==1]
data_not_fraud = data[data['Class']==0]
data_not_fraud_sample = data_not_fraud.sample(data_fraud.shape[0], replace=False, random_state=42)
data_balance = pd.concat([data_not_fraud_sample, data_fraud], axis=0).sample(frac=1, replace=False,random_state=101).reset_index().drop('index',axis=1)
```

```python
data_balance
```

```python
fraud_count = data_balance['Class'].value_counts()
fraud_rate = 100*fraud_count/data_balance.shape[0]
fraud_data = pd.concat([fraud_count,fraud_rate],axis=1).reset_index()
fraud_data.columns = ['class','Count','Percentage']
fraud_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_balance.drop('Class',axis=1), data_balance['Class'],test_size=0.2,random_state=101)
X_train.shape, y_train.shape
```

```python
randomForestModel = Pipeline([
    ('scaler',StandardScaler()),
    ('classifier',RandomForestClassifier())
])
randomForestModel.fit(X_train, y_train)
```

```python
y_prediction_logis = randomForestModel.predict(X_test)
y_prediction_logis
```

```python
cr = classification_report(y_test, y_prediction_logis)
print(cr)
```
